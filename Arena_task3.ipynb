```python
import pandas as pd

df = pd.read_csv("sales.csv")
print(df.head())
```

```python
print("Rows & Columns:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nSummary Info:")
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df = df.fillna(0)
```

```python
df = df.dropna()
```

```python
print("Average Price:", df['Price'].mean())
print("Highest Quantity:", df['Quantity'].max())
print("Lowest Quantity:", df['Quantity'].min())
```

```python
df['Total'] = df['Price'] * df['Quantity']
```

```python
print("Total Sales =", df['Total'].sum())
```

```python
best = df.groupby('Product')['Quantity'].sum().idxmax()
print("Best Selling Product:", best)
```

```python
print("\n----- SALES REPORT -----")
print("Total Sales:", df['Total'].sum())
print("Best Selling Product:", best)
print("\nSales by Product:")
print(df.groupby('Product')['Total'].sum())
```
